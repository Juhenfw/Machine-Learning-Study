```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score
# import tensorflow as tf
# import tensorflow.keras as keras
# from tensorflow.keras.layers import Dense, Input, Dropout
import matplotlib.pyplot as plt


headers = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'class']
df = pd.read_csv('D:/Materi Kuliah UNAIR/Semester 5/Pembelajaran Mesin (Praktikum) RK-A2/iris/iris.data',
                 header=None, names=headers)

# # Set pandas options to display all rows and columns
# pd.set_option('display.max_rows', None)   # Menampilkan semua baris
# pd.set_option('display.max_columns', None) # Menampilkan semua kolom

print(df)
```

### Setting 'Class' menjadi Target

```python
X = df.drop('class', axis=1)
y = df['class']

print(y)
```

```python
from sklearn.preprocessing import LabelEncoder

# Convert the 'class' column to numerical labels
le = LabelEncoder()
y = le.fit_transform(y)

print(y)
```

### Split untuk Training Data (70%), Test Data (30%)

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

### Training Decision Tree

```python
# Decision tree using Gini criterion
clf_gini = DecisionTreeClassifier(criterion='gini', random_state=42)
clf_gini.fit(X_train, y_train)

# Decision tree using Entropy criterion
clf_entropy = DecisionTreeClassifier(criterion='entropy', random_state=42)
clf_entropy.fit(X_train, y_train)
```

### Prediksi Menggunakan Decision Tree

```python
# Model Decision Tree menggunakan Gini
clf_gini = DecisionTreeClassifier(criterion='gini', random_state=42)
clf_gini.fit(X_train, y_train)

# Model Decision Tree menggunakan Entropy
clf_entropy = DecisionTreeClassifier(criterion='entropy', random_state=42)
clf_entropy.fit(X_train, y_train)

# Prediksi menggunakan model Gini
y_pred_gini = clf_gini.predict(X_test)

# Prediksi menggunakan model Entropy
y_pred_entropy = clf_entropy.predict(X_test)

# Confusion matrix untuk model Gini
conf_matrix_gini = confusion_matrix(y_test, y_pred_gini)

# Confusion matrix untuk model Entropy
conf_matrix_entropy = confusion_matrix(y_test, y_pred_entropy)

print("Confusion Matrix Gini:\n", conf_matrix_gini)
print("\nConfusion Matrix Entropy:\n", conf_matrix_entropy)
```

### Menghitung Confusion Matrix dan Metrik Evaluasi

```python
# Fungsi untuk menghitung metrik evaluasi
def calculate_metrics(conf_matrix, y_true, y_pred):
    # Akurasi
    accuracy = accuracy_score(y_true, y_pred)
    
    # Presisi (macro averaged untuk beberapa kelas)
    precision = precision_score(y_true, y_pred, average='macro')
    
    # Sensitivitas/Recall (macro averaged untuk beberapa kelas)
    sensitivity = recall_score(y_true, y_pred, average='macro')
    
    # Menghitung spesifisitas berdasarkan confusion matrix
    specificity = []
    for i in range(conf_matrix.shape[0]):
        tn = conf_matrix.sum() - (conf_matrix[i, :].sum() + conf_matrix[:, i].sum() - conf_matrix[i, i])
        fp = conf_matrix[:, i].sum() - conf_matrix[i, i]
        specificity.append(tn / (tn + fp))
    
    specificity = sum(specificity) / len(specificity)
    
    return accuracy, precision, sensitivity, specificity
```

### Hasil Evaluasi

```python
# Menghitung metrik untuk model Gini
metrics_gini = calculate_metrics(conf_matrix_gini, y_test, y_pred_gini)

# Menghitung metrik untuk model Entropy
metrics_entropy = calculate_metrics(conf_matrix_entropy, y_test, y_pred_entropy)

print("Metrics for Gini Decision Tree:")
print(f"Accuracy: {metrics_gini[0]*100:.2f}%")
print(f"Precision: {metrics_gini[1]*100:.2f}%")
print(f"Sensitivity (Recall): {metrics_gini[2]*100:.2f}%")
print(f"Specificity: {metrics_gini[3]*100:.2f}%")

print("\nMetrics for Entropy Decision Tree:")
print(f"Accuracy: {metrics_entropy[0]*100:.2f}%")
print(f"Precision: {metrics_entropy[1]*100:.2f}%")
print(f"Sensitivity (Recall): {metrics_entropy[2]*100:.2f}%")
print(f"Specificity: {metrics_entropy[3]*100:.2f}%")
```
