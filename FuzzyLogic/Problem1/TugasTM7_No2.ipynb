```python
def permintaan_turun(x):
    if(x<=500):
        nilai = 1
    elif(x>500 and x<2000):
        nilai = (2000-x)/(1500)
    elif(x>=2000):
        nilai = 0

    return nilai

def permintaan_sedang(x):
    if(x<=500):
        nilai = 0
    elif(x>500 and x<2000):
        nilai = (x-500)/(1500)
    elif(x>=2000 and x<=3500):
        nilai = 1
    elif(x>=3500 and x<=5000):
        nilai = (5000-x)/(1500)
    elif(x>=5000):
        nilai = 0

    return nilai

def permintaan_naik(x):
    if(x<=3500):
        nilai = 0
    elif(x>3500 and x<5000):
        nilai = (x-3500)/(1500)
    elif(x>=5000):
        nilai = 1

    return nilai

def persediaan_sedikit(x):
    if(x<=50):
        nilai = 1
    elif(x>50 and x<250):
        nilai = (250-x)/(200)
    elif(x>=250):
        nilai = 0

    return nilai

def persediaan_banyak(x):
    if(x<=50):
        nilai = 0
    elif(x>50 and x<250):
        nilai = (x-50)/(200)
    elif(x>=250):
        nilai = 1

    return nilai

def produksi_berkurang(x):
    if(x<=1000):
        nilai = 1
    elif(x>1000 and x<6000):
        nilai = (6000-x)/(5000)
    elif(x>=6000):
        nilai = 0

    return nilai

def produksi_bertambah(x):
    if(x<=1000):
        nilai = 0
    elif(x>1000 and x<6000):
        nilai = (x-1000)/(5000)
    elif(x>=6000):
        nilai = 1

    return nilai

def rule1(x,y):
  if (permintaan_turun(x)<=persediaan_banyak(y)):
    alpha = permintaan_turun(x)
  else:
    alpha = persediaan_banyak(y)

  z = 6000-(5000 * alpha)
  return alpha,z;

def rule2(x,y):
  if (permintaan_turun(x)<=persediaan_sedikit(y)):
    alpha = permintaan_turun(x)
  else:
    alpha = persediaan_sedikit(y)

  z = 6000-(5000 * alpha)
  return alpha,z;

def rule3(x,y):
  if (permintaan_naik(x)<=persediaan_banyak(y)):
    alpha = permintaan_naik(x)
  else:
    alpha = persediaan_banyak(y)

  z = 1000+(5000 * alpha)
  return alpha,z;

def rule4(x,y):
  if (permintaan_naik(x)<=persediaan_sedikit(y)):
    alpha = permintaan_naik(x)
  else:
    alpha = persediaan_sedikit(y)

  z = 1000+(5000 * alpha)
  return alpha,z;

def tsukamoto(x,y):
  alpha1,z1=rule1(x,y)
  alpha2,z2=rule2(x,y)
  alpha3,z3=rule3(x,y)
  alpha4,z4=rule4(x,y)
  alpha=(alpha1*z1)+(alpha2*z2)+(alpha3*z3)+(alpha4*z4)
  z=alpha1+alpha2+alpha3+alpha4
  return alpha/z

p=4000
q=500

print(f"Jika di ketahui Permintaan = {p} dan Persediaan = {q} maka Produksi = {tsukamoto(p,q)}")
```
