```python
import numpy as np

# Function to calculate final grade for each student
def calculate_final_grade(homework_scores, uts_score, uas_score):
    # Step 1: Calculate average homework score
    avg_homework_score = sum(homework_scores) / len(homework_scores)
    
    # Fuzzy sets for homework, UTS, and UAS
    fuzzy_low = np.array([0, 30, 60])
    fuzzy_medium = np.array([30, 70, 100])
    fuzzy_high = np.array([70, 100, 100])
    
    # Calculate membership values for homework, UTS, and UAS
    homework_low_membership = np.interp(avg_homework_score, fuzzy_low, [0, 1, 0])
    homework_medium_membership = np.interp(avg_homework_score, fuzzy_medium, [0, 1, 0])
    homework_high_membership = np.interp(avg_homework_score, fuzzy_high, [0, 1, 0])

    uts_low_membership = np.interp(uts_score, fuzzy_low, [0, 1, 0])
    uts_medium_membership = np.interp(uts_score, fuzzy_medium, [0, 1, 0])
    uts_high_membership = np.interp(uts_score, fuzzy_high, [0, 1, 0])

    uas_low_membership = np.interp(uas_score, fuzzy_low, [0, 1, 0])
    uas_medium_membership = np.interp(uas_score, fuzzy_medium, [0, 1, 0])
    uas_high_membership = np.interp(uas_score, fuzzy_high, [0, 1, 0])
    
    # Convert fuzzy membership values to scores
    homework_score = max(homework_low_membership * 1, homework_medium_membership * 2, homework_high_membership * 3)
    uts_score_conv = max(uts_low_membership * 1, uts_medium_membership * 2, uts_high_membership * 3)
    uas_score_conv = max(uas_low_membership * 1, uas_medium_membership * 2, uas_high_membership * 3)
    
    # Calculate the final fuzzy score
    final_fuzzy_score = (homework_score + uts_score_conv + uas_score_conv) / 3
    
    # Determine final grade based on fuzzy score
    if final_fuzzy_score < 1.5:
        final_grade = "C"
    elif 1.5 <= final_fuzzy_score < 2.5:
        final_grade = "B"
    else:
        final_grade = "A"
    
    return final_fuzzy_score, final_grade

# Calculate final grade for Budi
budi_homework_scores = [80, 85, 75]
budi_uts_score = 80
budi_uas_score = 40
budi_final_score, budi_final_grade = calculate_final_grade(budi_homework_scores, budi_uts_score, budi_uas_score)
print(f"Budi's final score: {budi_final_score:.2f}, Final Grade: {budi_final_grade}")

# Predict results for Budi's friends
friends_data = [
    {"name": "Jasmine", "homework_scores": [70, 80, 75], "uts_score": 80, "uas_score": 40},
    {"name": "Rafik", "homework_scores": [90, 80, 80], "uts_score": 77, "uas_score": 90},
    {"name": "Mutia", "homework_scores": [50, 55, 60], "uts_score": 10, "uas_score": 30},
    {"name": "Patrik", "homework_scores": [10, 0, 90], "uts_score": 60, "uas_score": 70},
    {"name": "Fajar", "homework_scores": [65, 35, 100], "uts_score": 70, "uas_score": 72}
]

for friend in friends_data:
    final_score, final_grade = calculate_final_grade(friend["homework_scores"], friend["uts_score"], friend["uas_score"])
    print(f"{friend['name']}'s final score: {final_score:.2f}, Final Grade: {final_grade}")
```

```python
import pandas as pd

# Fungsi untuk menghitung rata-rata tugas
def hitung_rata_rata_tugas(tugas):
    return sum(tugas) / len(tugas)

# Fungsi untuk menghitung hasil fuzzy berdasarkan keanggotaan nilai
def hitung_fuzzy(n_tugas, n_uts, n_uas):
    def konversi_nilai(nilai):
        if nilai < 60:
            return 1  # Low
        elif nilai < 80:
            return 2  # Middle
        else:
            return 3  # High

    tugas_konversi = konversi_nilai(n_tugas)
    uts_konversi = konversi_nilai(n_uts)
    uas_konversi = konversi_nilai(n_uas)
    
    # Menghitung nilai akhir fuzzy
    total_konversi = (tugas_konversi + uts_konversi + uas_konversi) / 3
    
    if total_konversi < 1.5:
        return "Low"
    elif 1.5 <= total_konversi < 2.5:
        return "Middle"
    else:
        return "High"

# Fungsi untuk mengonversi hasil fuzzy ke nilai akademik
def konversi_nilai_akhir(fuzzy_value):
    if fuzzy_value == "Low":
        return "C"
    elif fuzzy_value == "Middle":
        return "B"
    else:
        return "A"

# Data nilai Budi dan teman-temannya
data_nilai = {
    "Nama": ["Budi", "Jasmine", "Rafik", "Mutia", "Patrik", "Fajar"],
    "Tugas 1": [80, 70, 90, 50, 10, 65],
    "Tugas 2": [85, 80, 80, 55, 0, 35],
    "Tugas 3": [75, 75, 80, 60, 90, 100],
    "UTS": [80, 80, 77, 10, 60, 70],
    "UAS": [40, 40, 90, 30, 70, 72]
}

# Menghitung rata-rata tugas dan nilai fuzzy akhir
nilai_akhir = []
for i in range(len(data_nilai["Nama"])):
    tugas_avg = hitung_rata_rata_tugas([data_nilai["Tugas 1"][i], data_nilai["Tugas 2"][i], data_nilai["Tugas 3"][i]])
    fuzzy_value = hitung_fuzzy(tugas_avg, data_nilai["UTS"][i], data_nilai["UAS"][i])
    nilai_akhir.append(konversi_nilai_akhir(fuzzy_value))
    # print(tugas_avg)


print([hitung_rata_rata_tugas([data_nilai["Tugas 1"][i], data_nilai["Tugas 2"][i], data_nilai["Tugas 3"][i]]) for i in range(len(data_nilai["Nama"]))])

# Membuat DataFrame dengan format kolom "Tugas" yang terpisah
df_hasil = pd.DataFrame({
    "Nama": data_nilai["Nama"],
    "Tugas 1": data_nilai["Tugas 1"],
    "Tugas 2": data_nilai["Tugas 2"],
    "Tugas 3": data_nilai["Tugas 3"],
    "Rata-rata Tugas": [hitung_rata_rata_tugas([data_nilai["Tugas 1"][i], data_nilai["Tugas 2"][i], data_nilai["Tugas 3"][i]]) for i in range(len(data_nilai["Nama"]))],
    "UTS": data_nilai["UTS"],
    "UAS": data_nilai["UAS"],
    "Nilai Fuzzy": [hitung_fuzzy(hitung_rata_rata_tugas([data_nilai["Tugas 1"][i], data_nilai["Tugas 2"][i], data_nilai["Tugas 3"][i]]), data_nilai["UTS"][i], data_nilai["UAS"][i]) for i in range(len(data_nilai["Nama"]))],
    "Nilai Akhir Akademik": nilai_akhir
})

# Menampilkan hasil
df_hasil
```
