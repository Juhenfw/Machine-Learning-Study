```python
def layanan_buruk(x):
    nilai = 0
    if (x <= 5):
        nilai = 1
    elif (x > 5 and x < 6):
        nilai = (6 - x) / 1
    elif (x >= 6):
        nilai = 0
    
    return nilai

def layanan_baik(x):
    nilai = 0
    if (x < 5 or x > 9):
        nilai = 0
    elif (x >= 5 and x < 6):
        nilai = (x - 5) / 1
    elif (x >= 6 and x <= 8):
        nilai = 1
    elif (x > 8 and x < 9):
        nilai = (9 - x) / 1
    
    return nilai

def layanan_sangat_baik(x):
    nilai = 0
    if (x <= 8):
        nilai = 0
    elif (x > 8 and x < 9):
        nilai = (x - 8) / 1
    elif (x >= 9):
        nilai = 1
    
    return nilai

def makanan_buruk(x):
    nilai = 0
    if (x <= 5):
        nilai = 1
    elif (x >= 5 and x <= 6):
        nilai = (6 - x) / 1
    elif (x >= 6):
        nilai = 0
    
    return nilai

def makanan_layak(x):
    nilai = 0
    if (x < 5 or x > 9):
        nilai = 0
    elif (x >= 5 and x < 6):
        nilai = (x - 5) / 1
    elif (x >= 8 and x < 9):
        nilai = (9 - x) / 1
    elif (x >= 6 and x < 8):
        nilai = 1
    
    return nilai

def makanan_sangat_enak(x):
    nilai = 0
    if (x <= 6):
        nilai = 0
    elif (x > 8 and x < 9):
        nilai = (x - 8) / 1
    elif (x >= 9):
        nilai = 1
    
    return nilai

def tip_rendah(x):
    nilai = 0
    if (x <= 8):
        nilai = 1
    elif (x > 8 and x < 9):
        nilai = (9 - x) / 1
    else:
        nilai = 0
    
    return nilai

def tip_sedang(x):
    nilai = 0
    if (x < 8 or x > 18):
        nilai = 0
    elif (x >= 8 and x < 9):
        nilai = (x - 8) / 1
    elif (x >= 17 and x < 18):
        nilai = (18 - x) / 1
    elif (x >= 9 and x <= 17):
        nilai = 1
    
    return nilai

def tip_tinggi(x):
    nilai = 0
    if (x < 17):
        nilai = 0
    elif (x >= 17 and x <= 18):
        nilai = (x - 17) / 1
    elif (x >= 25):
        nilai = 1
    
    return nilai

# Aturan fuzzy
def rule1(layanan, makanan):
    alpha = max(layanan_sangat_baik(layanan), makanan_sangat_enak(makanan))
    z = 18 + (7 * alpha)  # Tip Tinggi
    return alpha, z

def rule2(layanan, makanan):
    alpha = layanan_baik(layanan)
    z = 9 + (8 * alpha)  # Tip Sedang
    return alpha, z

def rule3(layanan, makanan):
    alpha = min(layanan_buruk(layanan), makanan_buruk(makanan))
    z = 8 - (7 * alpha)  # Tip Rendah
    return alpha, z

# Metode Tsukamoto untuk prediksi
def tsukamoto(layanan, makanan):
    alpha1, z1 = rule1(layanan, makanan)
    alpha2, z2 = rule2(layanan, makanan)
    alpha3, z3 = rule3(layanan, makanan)
    
    numerator = (alpha1 * z1) + (alpha2 * z2) + (alpha3 * z3)
    denominator = alpha1 + alpha2 + alpha3
    
    if denominator == 0:
        return 0
    return numerator / denominator

# Prediksi untuk data
data = [
    (9.8, 6.8),
    (9.5, 9.2),
    (6.4, 9.6),
    (5.7, 6.3),
    (1.3, 2.5),
    (7.4, 6.8)
]

for i, (layanan, makanan) in enumerate(data):
    hasil_tip = tsukamoto(layanan, makanan)
    print(f"Data {i+1}: Kualitas Layanan = {layanan}, Kualitas Makanan = {makanan}, Prediksi Tip = {hasil_tip:.2f}%")
```
