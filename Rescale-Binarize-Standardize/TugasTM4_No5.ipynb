```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

auto_mpg = ['mpg', 'cylinders', 'displacement', 'horsepower', 'weight', 'acceleration', 'model_year', 'origin', 'car_name']

# Load dataset
df = pd.read_csv('D:/Materi Kuliah UNAIR/Semester 5/Pembelajaran Mesin (Praktikum) RK-A2/auto-mpg.data.csv',
                 sep = r'\s+', header = None, names = auto_mpg, na_values = '?', on_bad_lines='skip')

# Set pandas options to display all rows and columns
pd.set_option('display.max_rows', None)   # Menampilkan semua baris
pd.set_option('display.max_columns', None) # Menampilkan semua kolom

# # Print the DataFrame
# df
```

```python
for col in df.columns[:-1]:
    df[col] = pd.to_numeric(df[col], errors='coerce')
```

```python
print(df['horsepower'].isna().sum())
```

```python
from sklearn.preprocessing import MinMaxScaler 

array = df.values 
 
# Select column to rescale 
horsepower = array[:,3].reshape(-1,1)
 
# Rescaling data 
scaler = MinMaxScaler(feature_range=(0, 1)) 
rescaledX = scaler.fit_transform(horsepower) 

rescaledX
```

```python
from sklearn.preprocessing import Binarizer

df = df.dropna(subset=['horsepower'])  
array = df.values

# Binarize the data  
binarizer = Binarizer(threshold=100.0)  
binaryX = binarizer.fit_transform(array[:, 3].reshape(-1, 1)) 

binaryX
```

```python
from sklearn.preprocessing import StandardScaler

# preparation of dataframe using the data at given link and defined columns list 
array = df.values 
 
# Select column to rescale 
horsepower = array[:,3].reshape(-1,1)
 
# Standardize data 
scaler = StandardScaler().fit(horsepower) 
rescaledX = scaler.transform(horsepower) 

rescaledX
```
